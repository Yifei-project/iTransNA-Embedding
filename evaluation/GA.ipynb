### Lorenz   (y) (y+z)

```julia
for i = 1:100
    fid1 = h5open("../data/3D-lorenz/noisy_lorenz_no"*string(i)*".h5", "r")
    Y = fid1["xM"]
    if i==1
        w1 = estimate_delay(Y[:,1], "mi_min")
        w2 = estimate_delay(Y[:,2], "mi_min")
        w3 = estimate_delay(Y[:,3], "mi_min")
        w = maximum(hcat(w1, w2, w3))
        println(w)
    end
    close(fid1)
end
```
